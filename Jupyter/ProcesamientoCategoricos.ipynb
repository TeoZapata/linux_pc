```python
import pandas as pd
df = pd.read_csv("cars_26db3042-183f-4866-8b00-b607e8c2243b.csv")
df
```

```python
pd.get_dummies(df["engine_type"])
#esto es un one-hot
```

```python
import sklearn.preprocessing as ps

encoder= ps.OneHotEncoder(handle_unknown = "ignore")
```

```python
encoder.fit(df[["engine_type"]].values)#ajustarlos a las categorias de mi set
```

```python
encoder.transform([["gasoline"],["diesel"],["aceite"]]).toarray()
```

```python
encoder.fit(df[["year_produced"]].values)
```
